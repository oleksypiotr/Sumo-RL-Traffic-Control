```python
import os
import sys

if 'SUMO_HOME' in os.environ:
    sys.path.append(os.path.join(os.environ['SUMO_HOME'], 'tools'))
else:
    sys.exit("SUMO_HOME system env var not set")

from sumo_rl import SumoEnvironment
from stable_baselines3.dqn import DQN
from stable_baselines3.common.monitor import Monitor
```

```python
env = Monitor(
    SumoEnvironment(
        net_file="data/modlinska_swiatowida_kolacinska_net.net.xml",
        route_file="data/traffic_demand.rou.xml",
        delta_time = 10,
        single_agent=True,
        use_gui=False,
        # 7:00 - 9:00, real data
        begin_time=25200,
        num_seconds=7200,
        reward_fn="diff-waiting-time",
        ),
    filename="logs/dqn")
```

```python
"""
Possible reward functions:
reward_fns = {
        "diff-waiting-time": _diff_waiting_time_reward,
        "average-speed": _average_speed_reward,
        "queue": _queue_reward,
        "pressure": _pressure_reward,
        "co2": _co2_reward,
    }
https://github.com/LucasAlegre/sumo-rl/blob/main/sumo_rl/environment/traffic_signal.py
"""
```

```python
model = DQN(policy = 'MlpPolicy', env = env, learning_rate = 0.001, learning_starts = 72, verbose = 1, seed = 42)
```

```python
model.learn(total_timesteps=7200)
```

```python
model.save("models/dqn")
```
