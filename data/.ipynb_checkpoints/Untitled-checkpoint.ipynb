```python
import pandas as pd

try:
    df = pd.read_csv("traffic_modeled_24h_modlinska_swiatowida_kolacinska.csv")
except FileNotFoundError:
    print("BŁĄD: Nie znaleziono pliku CSV w tym folderze!")
    exit()

# ZAKTUALIZOWANY SŁOWNIK - Teraz wskazuje na najdalsze końce mapy (granice)
edge_mapping = {
    # Północ (od strony Jabłonny) - Startuje 215m od skrzyżowania
    'N_R': ('379408514#2', '143831377#0'),    # Modlińska -> Światowida
    'N_S': ('379408514#2', '179245142#0'),    # Modlińska -> Centrum
    'N_L': ('379408514#2', '178135939#0'),    # Modlińska -> Kołacińska
    'N_U': ('379408514#2', '178135941#0'),    # Zawrotka
    
    # Południe (od strony Centrum) - Startuje prawie 300m od skrzyżowania
    'S_R': ('179245145#4', '178135939#0'),    # Modlińska -> Kołacińska
    'S_S': ('179245145#4', '178135941#0'),    # Modlińska -> Jabłonna
    'S_L': ('179245145#4', '143831377#0'),    # Modlińska -> Światowida
    'S_U': ('179245145#4', '179245142#0'),    # Zawrotka
    
    # Zachód (Światowida) - Tutaj odcinek i tak stanowi granicę mapy
    'SW_R': ('-143831377#0', '179245142#0'),  # Światowida -> Centrum
    'SW_S': ('-143831377#0', '178135939#0'),  # Światowida -> Kołacińska
    'SW_L': ('-143831377#0', '178135941#0'),  # Światowida -> Jabłonna
    'SW_U': ('-143831377#0', '143831377#0'),  # Zawrotka
    
    # Wschód (Kołacińska) - Tutaj odcinek również stanowi granicę mapy
    'NE_R': ('-178135939#0', '178135941#0'),  # Kołacińska -> Jabłonna
    'NE_S': ('-178135939#0', '143831377#0'),  # Kołacińska -> Światowida
    'NE_L': ('-178135939#0', '179245142#0'),  # Kołacińska -> Centrum
    'NE_U': ('-178135939#0', '178135939#0')   # Zawrotka
}

with open("traffic_demand.rou.xml", "w", encoding="utf-8") as f:
    f.write('<?xml version="1.0" encoding="UTF-8"?>\n')
    f.write('<routes>\n')
    # Dodano departPos="base", by upewnić się, że auto rodzi się w punkcie 0 danego odcinka
    f.write('    <vType id="car" vClass="passenger" maxSpeed="14.0" accel="2.6" decel="4.5" length="5.0"/>\n\n')

    for idx, row in df.iterrows():
        begin_sec = idx * 900
        end_sec = (idx + 1) * 900
        
        for col, edges in edge_mapping.items():
            if col in df.columns:
                veh_count = int(row[col])
                if veh_count > 0:
                    from_e, to_e = edges
                    if from_e == to_e: continue 
                    flow_id = f"flow_{col}_{begin_sec}"
                    # Auta pojawią się na samym brzegu i znikną dopiero na drugim brzegu
                    f.write(f'    <flow id="{flow_id}" type="car" begin="{begin_sec}" end="{end_sec}" number="{veh_count}" from="{from_e}" to="{to_e}" departLane="best" departPos="base" departSpeed="max"/>\n')

    f.write('</routes>\n')

print("SUKCES! Wygenerowano plik z nowymi, skrajnymi krawędziami.")
```
